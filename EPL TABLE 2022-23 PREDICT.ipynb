## Predict New Season

```python
from sklearn.model_selection import train_test_split
```

```python
dataset
```

```python
dataset.shape
```

```python
target = dataset["target"]
```

```python
target[0:20]
```

```python
target.shape
```

```python
RS = np.arange(1,100,1)

train_results = []
test_results = []

for i in range(len(RS)): 
    x_train, x_test, y_train, y_test = train_test_split(dataset,
                                                        target,
                                                        test_size=0.2,
                                                        random_state= i)
    model = RandomForestClassifier(random_state = 0)
    model.fit(x_train[predictors],y_train)
     #Prediksi nilai y dari data train
    y_predt = model.predict(x_train[predictors])
#     #Prediksi nilai y dari data test
    y_pred = model.predict(x_test[predictors])
    
    train_acc = model.score(x_train[predictors],y_train)
    train_results.append(train_acc)
    
    test_acc = model.score(x_test[predictors],y_test)
    test_results.append(test_acc)
    print(i)
    print(f'Final Training Accuracy: {model.score(x_train[predictors],y_train)*100}%')
    print(f'Model Accuracy: {model.score(x_test[predictors],y_test)*100}%')
    
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
len(RS)
```

```python

from sklearn.metrics import precision_score, recall_score, accuracy_score, confusion_matrix
```
