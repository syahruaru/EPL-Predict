```python
alls["opp_code"] = alls["AwayTeam"].astype("category").cat.codes
```

```python
alls["tim_code"] = alls["HomeTeam"].astype("category").cat.codes
```

```python
alls["tim"] = alls["HomeTeam"]
```

```python
alls
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators= 100, min_samples_split=10, random_state=1)
```

```python
train = alls[alls["Date"]<'2022-01-01']
```

```python
test = alls[alls["Date"]>'2022-01-01']
```

```python
predictors = ["opp_code","FTHG","FTAG","tim_code"]
```

```python
predictors
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
error = accuracy_score(test["target"], preds)
```

```python
error
```

```python
combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))
```

```python
pd.crosstab(index=combined["actual"], columns=combined["predicted"])
```

```python
from sklearn.metrics import precision_score

precision_score(test["target"], preds, average='micro')
```
